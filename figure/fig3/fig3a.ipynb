```python
import pandas as pd
import numpy as np
```

```python
drugs_rank = pd.read_csv('../../map_drugs/drugs_rank.csv')
```

```python
drugs_rank.drop_duplicates('CHEMBL ID')
```

```python
clinicaltrial = pd.read_csv('../../map_drugs/clinicaltrial.csv')
clinicaltrial_all = clinicaltrial[(clinicaltrial['ChEMBL'].isin(drugs_rank['CHEMBL ID'].values))&(clinicaltrial['Clinical_phase']>=0)]['ChEMBL'].unique()
```

```python
drugs_candidate = drugs_rank[(drugs_rank.index<100)&(drugs_rank['Clinical Phase']>=4)]['CHEMBL ID'].values
```

```python
anno_reposition = pd.read_csv('../../map_drugs/anno_repostition.txt', sep='\t')
def mapping_score(path):
    if path == '':
        return 1
    else:
        return 1/len(path.split('-'))

anno_reposition['Mapping Score'] = anno_reposition['Path'].fillna('').apply(mapping_score)
```

```python
anno_reposition
```

```python
fig3a = anno_reposition[anno_reposition['CHEMBL ID'].isin(drugs_candidate)].copy()
fig3a['mapping'] = 'direct'
fig3a['mapping'] = fig3a['mapping'].where(fig3a['RWR Prob']==1, 'network')
fig3a = fig3a[['Drug','Virus','Host Gene','Drug Effect','Direction','Clinical Phase','CHEMBL ID','mapping','Mapping Score']]
fig3a = fig3a[fig3a['Mapping Score']>=0.5]
fig3a = fig3a.sort_values(['Virus','Host Gene','mapping'], ascending=[False, True, True])
fig3a = fig3a.drop_duplicates(['Virus','CHEMBL ID','Host Gene'])
fig3a['trial'] = 1
fig3a['trial'] = fig3a['trial'].where(fig3a['CHEMBL ID'].isin(clinicaltrial_all), 0)
fig3a['Drug'] = fig3a['Drug'].str.upper()
```

```python
fig3a = fig3a.drop_duplicates(['CHEMBL ID','Virus'])
fig3a['Drug'] = fig3a['Drug'].str.title()
fig3a['Drug'] = fig3a['Drug'].replace({'Acetylsalicylic Acid':'Aspirin','Byl719':'BYL719','Bay80-6946':'BAY80-6946'})
```

```python
fig3a_edge = pd.DataFrame(columns=['source', 'target', 'direction', 'mapping'])
ith = 0
for drug, virus, effct, mapping in fig3a[[
        'Drug', 'Virus', 'Drug Effect',
        'mapping'
]].values:
    fig3a_edge.loc[ith] = drug, virus, effct, mapping
    ith += 1
# for gene, virus, effct, mapping in fig3a[[
#         'Host Gene', 'Virus', 'Direction', 'mapping'
# ]].values:
#     fig3a_edge.loc[ith] = gene, virus, effct, 'direct'
#     ith += 1
fig3a_edge = fig3a_edge.drop_duplicates(['source','target'])
fig3a_edge.to_excel('./fig3a_edge_approved.xlsx',index=False)
```

```python
fig3a_node = pd.DataFrame(
    columns=['name', 'type', 'virus', 'max_phase', 'trial'])
ith = 0
for drug, virus, max_phase, trial in fig3a[[
        'Drug', 'Virus', 'Clinical Phase', 'trial'
]].values:
    fig3a_node.loc[ith] = drug, 'drug', virus, max_phase, trial
    ith += 1
# for gene, virus, max_phase, trial in fig3a[[
#         'Host Gene', 'Virus', 'Clinical Phase', 'trial'
# ]].values:
#     fig3a_node.loc[ith] = gene, 'gene', virus, np.nan, np.nan
#     ith += 1
for i, virus in enumerate(['SARS-CoV2', 'SARS-CoV', 'MERS-CoV'], start=1):
    fig3a_node.loc[ith] = virus, f'virus{i}', np.nan, np.nan, np.nan
    ith += 1
fig3a_node = fig3a_node.drop_duplicates(['name','type'])
fig3a_node.to_excel('./fig3a_node_approved.xlsx',index=False)
```
